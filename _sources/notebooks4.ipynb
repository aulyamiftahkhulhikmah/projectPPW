# **Perhitungan Manual PageRank**

PageRank adalah algoritma yang dikembangkan oleh Larry Page dan Sergey Brin, pendiri Google.Tujuannya adalah mengukur dan memberi nilai pada pentingnya sebuah halaman web berdasarkan jumlah dan kualitas tautan yang mengarah ke halaman tersebut. PageRank bekerja dengan memberikan skor pada setiap halaman web, yang ditentukan oleh seberapa banyak tautan yang mengarah ke halaman tersebut dari halaman-halaman lain. Halaman dengan lebih banyak tautan berkualitas tinggi mendapatkan skor PageRank yang lebih tinggi. Algoritma ini digunakan oleh mesin pencari, seperti Google, untuk menyusun hasil pencarian dan memberikan prioritas pada halaman-halaman yang dianggap lebih otoritatif.




## Rumus Perhitungan PageRank

Rumus dasar PageRank dapat dijelaskan secara sederhana sebagai berikut.

$$
PR(u) = (1-d) . \frac{1}{N} + d . \sum_{vâˆˆBu} \frac{PR(v)}{L(v)}
$$

* PR(u) : PageRank dari halaman (u)
* d : damping factor, biasanya diatur 0,85
* N : jumlah total halaman dalam graf (jumlah node dalam graf)
* B_u : himpunan halaman yang memiliki tautan ke halaman (u)
* L(v) : jumlah tautan keluar dari halaman (v)
